```python
def recommend_approach(scenario: dict) -> str:
    data_changes = scenario['data_changes_frequently']
    needs_format = scenario['need_specific_output_format']
    budget = scenario['budget']
    latency = scenario['latency_sensitive']
    knowledge = scenario['knowledge_type']

    # Prompt Engineering only - no new knowledge, no format needs, low budget
    if knowledge == 'behavioral' and budget == 'low' and not needs_format and not data_changes:
        return "Prompt Engineering only — task can be solved by better prompting, no training needed."

    # RAG + Fine-Tuning - needs both new facts AND specific format/behavior
    if knowledge == 'both' or (knowledge == 'factual' and needs_format):
        return "RAG + Fine-Tuning — needs both updated factual knowledge and behavioral/format adaptation."

    # Fine-Tuning - behavioral change or specific format needed, stable data, enough budget
    if (knowledge == 'behavioral' or needs_format) and budget in ['medium', 'high'] and not data_changes:
        return "Fine-Tuning — behavior or output format change needed and budget allows retraining."

    # RAG - frequently changing data or purely factual knowledge, no format needs
    if knowledge == 'factual' or data_changes:
        return "RAG — knowledge is factual or changes frequently; no retraining needed."

    return "Prompt Engineering only — no strong signal for RAG or Fine-Tuning."


# Testing 5 scenarios
scenarios = [
    {
        "data_changes_frequently": True,
        "need_specific_output_format": False,
        "budget": "low",
        "latency_sensitive": True,
        "knowledge_type": "factual"
    },
    {
        "data_changes_frequently": False,
        "need_specific_output_format": True,
        "budget": "high",
        "latency_sensitive": False,
        "knowledge_type": "behavioral"
    },
    {
        "data_changes_frequently": True,
        "need_specific_output_format": True,
        "budget": "high",
        "latency_sensitive": False,
        "knowledge_type": "both"
    },
    {
        "data_changes_frequently": False,
        "need_specific_output_format": False,
        "budget": "low",
        "latency_sensitive": False,
        "knowledge_type": "behavioral"
    },
    {
        "data_changes_frequently": False,
        "need_specific_output_format": False,
        "budget": "medium",
        "latency_sensitive": True,
        "knowledge_type": "factual"
    }
]

for i, scenario in enumerate(scenarios):
    print(f"Scenario {i+1}: {scenario['knowledge_type']} | budget={scenario['budget']} | changes={scenario['data_changes_frequently']}")
    print(f"→ {recommend_approach(scenario)}\n")
```
